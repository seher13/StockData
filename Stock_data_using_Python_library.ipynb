```python
import yfinance as yf
```

```python
apple = yf.Ticker('AAPL')
```

```python
apple_info = apple.info
```

```python
apple_info['country']
```

```python
apple_share_price_data = apple.history(period='max')
```

```python
apple_share_price_data
```

```python
apple_share_price_data['Date_column'] = apple_share_price_data.index 
```

```python
apple_share_price_data
```

```python
apple_share_price_data.reset_index(inplace=True)
```

```python
apple_share_price_data
```

```python
apple_share_price_date = apple_share_price_data.iloc[:,:8]
```

```python
apple_share_price_date
```

```python
dividends = pd.DataFrame(apple.dividends)
```

```python
dividends
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(15,7))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)



ax1.plot(apple_share_price_data.iloc[:,0], apple_share_price_data.iloc[:,1])
ax1.set_ylim(bottom=0)
ax1.set_title('Apple stock price')
ax1.set_xlabel('Year')
ax1.set_ylabel('USD')

ax2.plot(apple.dividends)
ax2.set_ylim(bottom=0)
```
